```python
from bs4 import BeautifulSoup
from bs4.element import ResultSet
import requests
import bs4
```

```python
urls = [
    'https://www.gazetadopovo.com.br/feed/rss/mundo.xml',
    'https://g1.globo.com/rss/g1/tecnologia/',
    'https://g1.globo.com/rss/g1/turismo-e-viagem/',
    'https://g1.globo.com/rss/g1/planeta-bizarro/'
    'https://g1.globo.com/rss/g1/pa/para/',
    'https://g1.globo.com/rss/g1/sp/ribeirao-preto-franca/',
    'https://www.noticiasaominuto.com.br/rss/ultima-hora',
    'https://www.noticiasaominuto.com.br/rss/tech'
]
```

```python
for url in urls:
    response = requests.get(url)
    html = response.text
    site = BeautifulSoup(html, 'html.parser')
    titulos = site.findAll('title')
    print(url)
    itens = site.findAll('item')
    for item in itens:
        soup_cleaned = BeautifulSoup(item.description.get_text(), 'html.parser')
        final_text = soup_cleaned.get_text()
        print(item.title.text)
        print(item.url.text)
        print(item.guid.text)
        print(final_text)
        print(item.pubdate.text)
        print()
```
